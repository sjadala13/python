```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot([0, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('Y numbers')
plt.xlabel('X numbers')
plt.title('Simple plot')
```

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot. 
The default format string is ‘b-‘, which is a solid blue line.

```python
x = [1,2,3,4]
y =[1,4,9,16]
plt.plot(x, y, linewidth=4.0)
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0, 6, 0, 20])
```

```python
x = [1,2,3,4]
y =[1,4,9,16]
plt.plot(x, y, '_')
```

```python
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]

plt.plot(x, y, linestyle="dashed", marker="o", color="red")
```

```python
import seaborn as sns
import pandas as pd
```

```python
#http://stattrek.com/descriptive-statistics/variables.aspx?Tutorial=AP
```

```python
df = pd.read_csv('../data/seaborn_data.csv', index_col=0)
```

```python
df.head()
```

```python
#scatter plot
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
# Scatterplot arguments
#http://stattrek.com/statistics/charts/scatterplot.aspx?Tutorial=AP
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # No regression line
           hue='Stage')   # Color by evolution stage
```

```python
# Plot using Seaborn
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, 
           hue='Stage')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Boxplot
sns.boxplot(data=df)
#http://stattrek.com/statistics/charts/boxplot.aspx?Tutorial=AP
```

```python
# Pre-format DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
 
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')

# Violin plot
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
#Violin plots are useful alternatives to box plots.
#They show the distribution (through the thickness of the violin) instead of only the summary statistics.
```

```python
sns.swarmplot(x='Type 1', y='Attack', data=df)
```

```python
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               )
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type')
plt.xticks(rotation=-45)
```

```python
#Heatmaps help you visualize matrix-like data.

# Calculate correlations
corr = stats_df.corr()
 
# Heatmap
sns.heatmap(corr)
```

```python
corr
```

```python
sns.heatmap(corr,annot = True)
```

```python
#Histogram
#Histograms allow you to plot the distributions of numeric variables.

sns.distplot(df.Attack)
```

```python
#Bar plots help you visualize the distributions of categorical(qualitative) variables.

sns.countplot(x='Type 1', data=df)
 
# Rotate x-labels
plt.xticks(rotation=-45)
```

```python
#Difference between bar charts and histograms
#With bar charts, each column represents a group defined by a categorical variable
#and with histograms, each column represents a group defined by a continuous, quantitative variable.
```

```python
#Factor plots make it easy to separate plots by categorical classes.

g = sns.factorplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45)
```

```python
#Joint distribution plots combine information from scatter plots and histograms to give you detailed information for bi-variate distributions
sns.jointplot(x='Attack', y='Defense', data=df)
```
